```python
from math import e

x = 0.45
h = 0.005
xh = 0.455

fx = e**(-x)
fxh = 0
fxd = (fx)*-1

fxh1=fx+fxd*h

def factorial(n):
   if n==0 or n==1:
            resultado=1
   elif n>1:
            resultado=n*factorial(n-1)
   return resultado




print("Orden 0 =",fx)
print("")
print("Orden 1 =",fxh1)
print("")
Vant = fxh1

for i in range(2,16):
  
  if i != 2:
    Vant = fxh
  


  print("Valor anterior = ",Vant)
  fxd = fxd*-1
  
  
  fxh = Vant + (fxd/factorial(i)) * (h**i)
  
  Vact = fxh
  
  mE=abs((Vact - Vant)/Vact)*1
  
  
  
  print("Valor actual = ",Vact)
  
  print("Orden",i, "=",fxh)
  print("Error aproximado =", mE)
  print("")

```
